### 03) WAP to demonstrate the use of Bar chart.

```python
left = [1,2,3,4,5]
height = [10,20,60,44,37]
c = ['b','g','r','c','m']
plt.title('Paresh')
plt.bar(left, height, color = c)
for i in range(len(height)):
    plt.text(left[i],height[i],height[i])
plt.show()
```

### 04) WAP to demonstrate the use of Histogram.

```python
import random
cpi = []
for i in range(101):
    st = random.randint(1,10)
    cpi.append(st)
plt.hist(cpi,bins=10,histtype='bar',rwidth=0.5)
plt.title("Hist Demo")
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import random
cpi = []
for i in range(101):
    st = random.randint(1,10)
    cpi.append(st)
counts,bins,patches = plt.hist(cpi,bins=10,histtype='bar',rwidth=0.5,alpha=0.7)
for count,x in zip(counts,bins):
    plt.text(x+0.3,count,str(int(count)),va='bottom')
plt.title("Hist Demo")
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
import pandas as pd
df = pd.read_csv('insurance.csv')
grouped = df.groupby(['smoker'])
for key,group in grouped:
    plt.scatter(group['age'],group['charges'])

plt.legend()
plt.show()
```
